```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

obiwan_elg = np.loadtxt('/global/homes/h/huikong/obiwan_test/obiwan_corr/corrfunc_data/NGC_Feb15.out').transpose()
johan_elg = np.loadtxt('/global/homes/h/huikong/obiwan_test/obiwan_corr/corrfunc_data/Johan_random.txt').transpose()
plt.gca().set_xscale('log')
plt.gca().set_yscale('log')
(_, caps, _) = plt.errorbar(obiwan_elg[0],obiwan_elg[1],obiwan_elg[2],marker='.',fillstyle='none',linestyle='none',capsize=5,color='dodgerblue')
for cap in caps:
    cap.set_markeredgewidth(2)
(_, caps, _) = plt.errorbar(johan_elg[0],johan_elg[1],johan_elg[2],marker = '.',fillstyle='none',linestyle='none',capsize=5,color='coral')
for cap in caps:
    cap.set_markeredgewidth(2)
#plt.gca().set_xlim((0.007, 1))
#plt.gca().set_ylim((0, 0.5))

np.set_printoptions(precision=3)

logA = np.log10(obiwan_elg[0])
logB = np.log10(obiwan_elg[1])
coefficients = np.polyfit(logA, logB, 1)
print 'obiwan:'+str(coefficients)
B = 10**coefficients[1]*np.power(obiwan_elg[0],coefficients[0])
obiwan, = plt.plot(obiwan_elg[0],B,color='dodgerblue',label = 'subset_Feb15 '+str("%0.2f" % coefficients[0]))

logA = np.log10(johan_elg[0])
logB = np.log10(johan_elg[1])
coefficients = np.polyfit(logA, logB, 1)
print 'johan:'+str(coefficients)
B = 10**coefficients[1]*np.power(johan_elg[0],coefficients[0])
johan, = plt.plot(johan_elg[0],B,color='coral',label = 'johan '+str("%0.2f" % coefficients[0]))
plt.xlabel(r'$\theta$'+'(Degree)')
plt.ylabel('w('+r'$\theta$'+')')
plt.title('ELG NGC VS Obiwan')
plt.legend(handles=[obiwan, johan])
```
